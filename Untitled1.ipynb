```python
import os
import shutil
import zipfile
import pandas as pd
import cv2
import glob
from sklearn.model_selection import train_test_split
from ultralytics import YOLO

# --- CONFIGURATION ---
ZIP_FILE = "/image 2.zip"
CSV_FILE = "/data.csv"
FINAL_DATA_PATH = "/content/yolo_dataset"

# Adjust these class names to match your annotation file exactly
CLASS_MAP = {"WBC": 0, "RBC": 1}

# --- 1. SETUP & DATA EXTRACTION ---
if os.path.exists(FINAL_DATA_PATH):
    shutil.rmtree(FINAL_DATA_PATH)
for split in ['train', 'val']:
    os.makedirs(f"{FINAL_DATA_PATH}/images/{split}", exist_ok=True)
    os.makedirs(f"{FINAL_DATA_PATH}/labels/{split}", exist_ok=True)

TEMP_EXTRACT_DIR = "/content/temp_extract"
if os.path.exists(TEMP_EXTRACT_DIR):
    shutil.rmtree(TEMP_EXTRACT_DIR)
os.makedirs(TEMP_EXTRACT_DIR, exist_ok=True)

with zipfile.ZipFile(ZIP_FILE, 'r') as zip_ref:
    zip_ref.extractall(TEMP_EXTRACT_DIR)

all_image_paths = []
for ext in ['*.jpg', '*.jpeg', '*.png', '*.JPG', '*.JPEG', '*.PNG']:
    all_image_paths.extend(glob.glob(os.path.join(TEMP_EXTRACT_DIR, '**', ext), recursive=True))

# --- 2. ANNOTATION PROCESSING & SPLIT ---
try:
    df = pd.read_csv(CSV_FILE, header=None)
    df.columns = ["filename", "class", "xmin", "ymin", "xmax", "ymax"]
except Exception:
    raise ValueError("Error reading CSV file. Check that the path is correct and it has 5 columns (filename, class, xmin, ymin, xmax, ymax).")

grouped = df.groupby('filename')
valid_paths = []
for path in all_image_paths:
    if os.path.basename(path) in grouped.groups:
        valid_paths.append(path)

train_paths, val_paths = train_test_split(valid_paths, test_size=0.2, random_state=42)

def process_data(paths, split):
    for src in paths:
        filename = os.path.basename(src)
        img = cv2.imread(src)
        if img is None: continue

        h, w = img.shape[:2]
        lines = []

        for _, row in grouped.get_group(filename).iterrows():
            if row['class'] not in CLASS_MAP: continue
            cls = CLASS_MAP[row['class']]
            bw = (row['xmax'] - row['xmin']) / w
            bh = (row['ymax'] - row['ymin']) / h
            xc = (row['xmin'] + row['xmax']) / 2.0 / w
            yc = (row['ymin'] + row['ymax']) / 2.0 / h
            lines.append(f"{cls} {xc:.6f} {yc:.6f} {bw:.6f} {bh:.6f}")

        with open(f"{FINAL_DATA_PATH}/labels/{split}/{os.path.splitext(filename)[0]}.txt", 'w') as f:
            f.write('\n'.join(lines))
        shutil.copy(src, f"{FINAL_DATA_PATH}/images/{split}/{filename}")

process_data(train_paths, 'train')
process_data(val_paths, 'val')

yaml_content = f"""
path: {FINAL_DATA_PATH}
train: images/train
val: images/val
nc: {len(CLASS_MAP)}
names: {list(CLASS_MAP.keys())}
"""

yaml_path = f"{FINAL_DATA_PATH}/data.yaml"
with open(yaml_path, "w") as f:
    f.write(yaml_content)

# --- 3. MODEL TRAINING (TWO VARIANCES) ---


model_v8 = YOLO("yolov8n.pt")
model_v8.train(
    data=yaml_path,
    epochs=20,
    imgsz=640,
    batch=8,
    project="/content/runs/detect",
    name="YOLOv8_train"
)

print(" Training YOLOv5s...")
model_v5 = YOLO("yolov5s.pt")
model_v5.train(
    data=yaml_path,
    epochs=20,
    imgsz=640,
    batch=8,
    project="/content/runs/detect",
    name="YOLOv5_train"
)

# --- 4. METRICS (ACCURACY) AND PLOTTING EVALUATION ---



v8_weights = "/content/runs/detect/YOLOv8_train/weights/best.pt"
v5_weights = "/content/runs/detect/YOLOv5_train/weights/best.pt"
val_source = f"{FINAL_DATA_PATH}/images/val"

# --- YOLOv8 EVALUATION ---
final_model_v8 = YOLO(v8_weights)
print("\n--- YOLOv8n Metrics ---")
metrics_v8 = final_model_v8.val(data=yaml_path)


final_model_v8.predict(
    source=val_source,
    save=True,
    conf=0.25,
    project="/content/runs/detect",
    name="YOLOv8_plots"
)

# --- YOLOv5 EVALUATION ---
final_model_v5 = YOLO(v5_weights)
print("\n--- YOLOv5s Metrics ---")
metrics_v5 = final_model_v5.val(data=yaml_path)

final_model_v5.predict(
    source=val_source,
    save=True,
    conf=0.25,
    project="/content/runs/detect",
    name="YOLOv5_plots"
)

```
